```python
df["gender"]=="female"
```

```python
df[df["class"]=="Four"][["name", "class", "gender"]]
```

```python
#df.loc[rows, names of columns]
```

```python
df.loc[df["class"]=="Four", ["name", "class", "gender"]]
```

```python
# df.iloc[df["class"]=="Four",[1, 2, 4]]
# iloc doesnt support filtering
```

- Write code to return name, class, gender of the male student at class Four

- when using multiple columns to filter, enclose each condition by () and separate them by &

```python
df[(df["class"]=="Four") & (df["gender"]=="male")]
```

- What are the names, class of female student who has mark less than 70?

```python
df[(df["mark"]<70) & (df["gender"]=="female")][["name", "class"]]
# & operator means both conditions must be met
# both conditon must be true
```

```python
df[(df["mark"]<70) | (df["gender"]=="female")][["name", "class", "gender", "mark"]]
# | or. one of the conditions must be true
```

```python
df.loc[ (df["mark"]<70) & (df["gender"]=="female"), ["name", "class"]]
```

```python
df[ (df["mark"]<70) & (df["gender"]=="female")]
```

# Sorting results

- There is always natural order of the rows. student info sorted by their *id*

```python
df.head()
```

```python
df.sort_values(by="name")
# alphabetical order
```

```python
df.sort_values(by="mark").head()
# ascending order
```

```python
df.sort_values(by="name", ascending=False).head(10)
# descending order
# ascending false will reverse the order. its in descending order
```

```python
df.sort_values(by="mark", ascending=False).head()
# ascending order
```

```python
# applyting sort after slicing
df[df["class"]=="Six"].sort_values(by="mark")
```

```python
df.loc[df["class"]=="Six", :].sort_values(by="mark")
```
